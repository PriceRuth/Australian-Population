It seems that there is a lot of data with different categories. We will start first with looking at Adelaide population by gender.

```python
#Narrowing down the data by selected rows we want to analyze
options = ['Estimated resident population - males (no.)',
       'Estimated resident population - females (no.)']

#filtering rows
pop_df = adelaide[adelaide['Description'].isin(options)]
pop_df
```

```python
#For Easier Visualization, we want to transpose the data where the year will be our index
pop_df1 = pop_df.transpose()
pop_df1
```

```python
#Deleting rows we don't need
#Click three times
pop_df1.drop(pop_df1.index[0], inplace=True)
pop_df1
```

```python
#Setting the "Year" and "Description" as our index
pop_df2 = pop_df1.reset_index().rename(columns={'index':'Year', 2: 'Estimated Male Residents', 3: 'Estimated Female Residents' })
pop_df2
```

```python
#Setting our x axis to "Year"
pop_df2 = pop_df2.set_index('Year')
pop_df2
```

```python
#Checking if we can visualize it now
pop_df2.plot()
```

Female population is much higher than male population in South Australia. It seems like that in 2020, the population is increasing at a slower rate (not massive) in both men and women.

Let's have a look at the population growth by number according to age

```python
#This is the group of rows we want to look up
options = [ 'Persons - 10-14 years (no.)', 'Persons - 15-19 years (no.)',
       'Persons - 20-24 years (no.)', 'Persons - 25-29 years (no.)',
       'Persons - 30-34 years (no.)', 'Persons - 35-39 years (no.)',
       'Persons - 40-44 years (no.)', 'Persons - 45-49 years (no.)',
       'Persons - 50-54 years (no.)', 'Persons - 55-59 years (no.)',
       'Persons - 60-64 years (no.)', 'Persons - 65-69 years (no.)',
       'Persons - 70-74 years (no.)', 'Persons - 75-79 years (no.)',
       'Persons - 80-84 years (no.)', 'Persons - 85 and over (no.)',]

#filtering rows
pop_by_number = adelaide[adelaide['Description'].isin(options)]
pop_by_number
```

```python
#We want to transpose and set "Description" as our index
pop_by_number = pop_by_number.set_index('Description').transpose()
pop_by_number
```

```python
#Remove some rows we don't need for visualization
#Click twice
pop_by_number.drop(pop_by_number.index[0], inplace=True)
pop_by_number
```

```python
#Let's try to drop columns that have no data
pop_by_number = pop_by_number.dropna()
pop_by_number
```

```python
#We will rename the year column and set it as our index
pop_by_number = pop_by_number.reset_index().rename(columns={'index':'Year'}).set_index('Year')
pop_by_number
```

```python
pop_by_number.plot(kind="bar", figsize=(10,6), align="center").legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

Let's see if we can get a more accurate result if we look up on population by percentage.

```python
#This is the group of rows we want to look up
#Population by percentage rate
options = [ 'Persons - 0-4 years (%)', 'Persons - 5-9 years (%)',
       'Persons - 10-14 years (%)', 'Persons - 15-19 years (%)',
       'Persons - 20-24 years (%)', 'Persons - 25-29 years (%)',
       'Persons - 30-34 years (%)', 'Persons - 35-39 years (%)',
       'Persons - 40-44 years (%)', 'Persons - 45-49 years (%)',
       'Persons - 50-54 years (%)', 'Persons - 55-59 years (%)',
       'Persons - 60-64 years (%)', 'Persons - 65-69 years (%)',
       'Persons - 70-74 years (%)', 'Persons - 75-79 years (%)',
       'Persons - 80-84 years (%)', 'Persons - 85 and over (%)']

#filtering rows
pop_by_perc = adelaide[adelaide['Description'].isin(options)]
pop_by_perc
```

```python
#We want to transpose and set "Description" as our index
pop_by_perc = pop_by_perc.set_index('Description').transpose()
pop_by_perc
```

```python
#Remove some rows we don't need for visualization
#Click twice
pop_by_perc.drop(pop_by_perc.index[0], inplace=True)
pop_by_perc
```

```python
#Let's try to drop columns that have no data
pop_by_perc = pop_by_perc.dropna()
pop_by_perc
```

```python
#Set the year column as our index
pop_by_perc = pop_by_perc.reset_index().rename(columns={'index':'Year'}).set_index('Year')
pop_by_perc
```

```python
#Transposing the table for visualization use
pop_by_perc = pop_by_perc.transpose()
pop_by_perc
```

```python
#Create a for loop bar plot to easily visualize population rate for each age category
for col in pop_by_perc.columns:
    pop_by_perc.plot.bar(y=col, figsize=(6,2)).legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
